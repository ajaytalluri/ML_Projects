```python
state_list = df['Area_Name'].drop_duplicates().to_list()
```

```python
state_list=df["Area_Name"].drop_duplicates().to_list()

df_major_reason_kidnap=pd.DataFrame(columns=["AREA","Major_reason_of_kidnap","Count_of_kidnaps_due_to_major_reason","Total_kidnaps","Proportion_of_kidnaps_major_reason"])
area_list=list()
major_reason_list=list()
count_kidnaps_major=list()
total_kidnaps=list()
proportion_list=list()
for state in state_list:
    max_kid=0
    for index in df_group.loc[state].index:
        
        if (max_kid<df_group.loc[state].loc[index][0]) and (index!="Kidnap - Total"):
            max_kid=df_group.loc[state].loc[index][0]
            index_final=index
            
    print("AREA: {} \nMajority cases due to: {} and count is: {}".format(state,index_final,max_kid))
    print("Ratio of cases due to {} is :{}".format(index_final,max_kid/df_group.loc[state].loc["Kidnap - Total"][0]))
    print("\n")
    area_list.append(state)
    major_reason_list.append(index_final)
    count_kidnaps_major.append(max_kid)
    total_kidnaps.append(df_group.loc[state].loc["Kidnap - Total"][0])
    proportion_list.append(max_kid/df_group.loc[state].loc["Kidnap - Total"][0])
```

```python
df_major_reason_kidnap["AREA"]=area_list
df_major_reason_kidnap["Major_reason_of_kidnap"]=major_reason_list
df_major_reason_kidnap["Count_of_kidnaps_due_to_major_reason"]=count_kidnaps_major
df_major_reason_kidnap["Total_kidnaps"]=total_kidnaps
df_major_reason_kidnap["Proportion_of_kidnaps_major_reason"]=proportion_list
df_major_reason_kidnap
```

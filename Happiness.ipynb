#### Univariate Analysis

```python
# selecting the numerical data for univariate analysis
df_cut =df.iloc[:,2:]
```

```python
df_cut.head()
```

```python
ncols=10
nrows=10
coll_list = df_cut.columns.values

plt.figure(figsize=(20,80))
for i in range(0,len(coll_list)):
    plt.subplot(nrows,ncols,i+1)
    sns.boxplot(df_cut[coll_list[i]], color ='green', orient='v')
    plt.tight_layout()
```

```python
ncols=10
nrows=10
coll_list = df_cut.columns.values

plt.figure(figsize=(40,80))
for i in range(0,len(coll_list)):
    plt.subplot(nrows,ncols,i+1)
    sns.distplot(df_cut[coll_list[i]], color ='green')
    plt.tight_layout()
```
